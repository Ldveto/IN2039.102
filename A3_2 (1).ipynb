```python
path = 'datasets/adult.csv'
df = pd.read_csv(path)
df.head(2)
```

```python
x = df.columns.tolist()
x.sort()
x
```

```python
myPivot = pd.pivot_table(df,
                         index = ['education','marital-status','sex'],
                         values = ['age','hours-per-week'],
                         aggfunc = ['mean'],
                         #aggfunc en python average es el default en pivot table y en excel es la suma
                        
                        )

pd.set_option('display.max_rows',5000)

myPivot
```

```python
myPivot = pd.pivot_table(df,
                         index = ['education','marital-status','sex'],
                         values = ['age','hours-per-week'],
                         aggfunc = ['median'],
                         #aggfunc en python average es el default en pivot table y en excel es la suma
                        
                        )

pd.set_option('display.max_rows',5000)

myPivot
```
